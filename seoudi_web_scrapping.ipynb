```python
def percentage(num1, num2):
    difference = num1 - num2
    percentage = (difference / num1) * 100
    return percentage
```

```python
def remove_string(barcode, input_str):
    try:
      if barcode in input_str:
        input_str = input_str.replace("-"+barcode, '')

      substrings = input_str.rsplit("-", 1)
      first_substring = substrings[0].replace("-", " ")
      new_string = first_substring + "-" + substrings[1]
      return new_string
    except Exception as e:
      if barcode in input_str:
        input_str = input_str.replace("-"+barcode, '')
      return input_str
```
